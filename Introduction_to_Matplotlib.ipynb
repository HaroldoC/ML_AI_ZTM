## Matplotlib Example Workflow

```python
# 0. import matplotlib and get it ready for plotting in jupyter
%matplotlib inline
import matplotlib.pyplot as plt

# 1. Prepare Data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. Setup Plot
fig, ax = plt.subplots(figsize=(10, 10)) # (width, height)

# 3. Plot Data
ax.plot(x, y)

# 4. Customize Plot
ax.set(title="Simple Plot",
       xlabel="x-axis",
       ylabel="y-axis")

# 5. Save & Show (you save the whoile figure)
fig.savefig("images/sample-plot.png")
```

## Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10, 5))
# Plot to each different axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```

```python
# Subplot option 2
fig, ax = plt.subplots(nrows=2,
                       ncols=2,
                       figsize=(10, 5));
# Plot to each different index
ax[0, 0].plot(x, x/2);
ax[0, 1].scatter(np.random.random(10), np.random.random(10));
ax[1, 0].bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax[1, 1].hist(np.random.randn(1000));
```

### Which one should be used? pyplot vs matplotlib OO method?
* When plotting something quickly, okay to use the pyplot method
* When plotting something more advanced, use the OO method 

```python
over_50 = heart_disease[heart_disease["age"] > 50]
over_50.head()
```

```python
# Pyplot method
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target");
```

```python
# OO method mixed with Pyplot method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            ax=ax);
ax.set_xlim([45, 100]);
```

```python
over_50.target.values
```

```python
## OO Method from scratch
fig, ax = plt.subplots(figsize=(10, 6))

# Plot the Data
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"])

# Customize the plot
ax.set(title="Heart Disease and Cholesterol Levels",
      xlabel="age",
      ylabel="chol")

# Add a legend
ax.legend(*scatter.legend_elements(), title="Target")

# Add horizontal line in OO method
ax.axhline(over_50["chol"].mean(),
          linestyle="--");
```

```python
# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10, 10),
                              sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"])

# Customize the plot
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="chol")

# Add a legend
ax0.legend(*scatter.legend_elements(), title="Target")

# Add horizontal line in OO method
ax0.axhline(over_50["chol"].mean(),
          linestyle="--");


# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"])

# Customize the plot
ax1.set(title="Heart Disease and Max Heart Rate",
      xlabel="age",
      ylabel="thalach")

# Add a legend
ax1.legend(*scatter.legend_elements(), title="Target")

# Add horizontal line in OO method
ax1.axhline(over_50["thalach"].mean(),
          linestyle="--")

# Add a tittle to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```

```python
# Set a new style
plt.style.use("seaborn-v0_8-poster")

# Subplot of chol, age, thalach
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10, 10),
                              sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50["age"],
                      y=over_50["chol"],
                      c=over_50["target"],
                      cmap="cool")

# Customize the plot
ax0.set(title="Heart Disease and Cholesterol Levels",
        ylabel="chol")

# Change the x axis limit
ax0.set_xlim([50, 80])

# Add a legend
ax0.legend(*scatter.legend_elements(), title="Target")

# Add horizontal line in OO method
ax0.axhline(over_50["chol"].mean(),
          linestyle="--");


# Add data to ax1
scatter = ax1.scatter(x=over_50["age"],
                      y=over_50["thalach"],
                      c=over_50["target"],
                      cmap="spring")

# Customize the plot
ax1.set(title="Heart Disease and Max Heart Rate",
      xlabel="age",
      ylabel="thalach")

# Change the x axis limit
ax1.set_xlim([50, 80])
ax1.set_ylim([60, 200])

# Add a legend
ax1.legend(*scatter.legend_elements(), title="Target")

# Add horizontal line in OO method
ax1.axhline(over_50["thalach"].mean(),
          linestyle="--")

# Add a tittle to the figure
fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight="bold");
```
